```python
## import visualization packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test = pd.read_csv("test.csv")
test.head()
```

```python
test.info()
```

```python
## We are missing several values from age and 1 from Fare
```

```python
## View the missing data for fare (should be quick to handle)
missing = test.loc[test.Fare.isnull(),:]
missing
```

```python
## replace with the median fare
test.Fare = test.Fare.replace(np.NaN, test.loc[test.Pclass == 3, 'Fare'].median())
test.info()
```

```python
## View the missing age 
missing = test.loc[test.Age.isnull(), :]
missing
```

```python
## Replace missing age values with the median from their pclass and sex

## Male
test.loc[(test.Pclass == 1) & (test.Sex == 'male'), 'Age'] = test.loc[(test.Pclass == 1) & (test.Sex == 'male'), 'Age'].replace(np.NaN, test.loc[(test.Pclass == 1) & (test.Sex == 'male'), 'Age'].median())
test.loc[(test.Pclass == 2) & (test.Sex == 'male'), 'Age'] = test.loc[(test.Pclass == 2) & (test.Sex == 'male'), 'Age'].replace(np.NaN, test.loc[(test.Pclass == 2) & (test.Sex == 'male'), 'Age'].median())
test.loc[(test.Pclass == 3) & (test.Sex == 'male'), 'Age'] = test.loc[(test.Pclass == 3) & (test.Sex == 'male'), 'Age'].replace(np.NaN, test.loc[(test.Pclass == 3) & (test.Sex == 'male'), 'Age'].median())

## Female
test.loc[(test.Pclass == 1) & (test.Sex == 'female'), 'Age'] = test.loc[(test.Pclass == 1) & (test.Sex == 'female'), 'Age'].replace(np.NaN, test.loc[(test.Pclass == 1) & (test.Sex == 'female'), 'Age'].median())
test.loc[(test.Pclass == 2) & (test.Sex == 'female'), 'Age'] = test.loc[(test.Pclass == 2) & (test.Sex == 'female'), 'Age'].replace(np.NaN, test.loc[(test.Pclass == 2) & (test.Sex == 'female'), 'Age'].median())
test.loc[(test.Pclass == 3) & (test.Sex == 'female'), 'Age'] = test.loc[(test.Pclass == 3) & (test.Sex == 'female'), 'Age'].replace(np.NaN, test.loc[(test.Pclass == 3) & (test.Sex == 'female'), 'Age'].median())
```

```python
test.info()
```

## Feature Engineering

```python
## Create prefix feature


## Split the name into componenents and keep the prefix 
intermediate = test["Name"].str.split(pat = ".", n = 1, expand = True)
prefix = intermediate[0].str.split(",", expand = True)
prefix.head()

test['Prefix'] = prefix[1].str.lstrip()

test.Prefix.unique()
```

```python
## Create same categories as training set (Mr, Mrs, Ms, Honor)

test.Prefix = test.Prefix.replace(['Col', 'Rev', 'Dr', 'Dona'], 'Honor')
test.Prefix = test.Prefix.replace('Miss', 'Ms')
test.Prefix = test.Prefix.replace('Master', 'Mr')
test.Prefix.unique()
```

```python
## Create family size
test['Family_size'] = test['SibSp'] + test['Parch']
test.head()
```

```python
## Create sex * pclass categories

## Need to get dummy values first
test = pd.get_dummies(test, columns = ['Sex'],  drop_first = True)

## Create cateogies A - F
test['Sex_pclass'] = (test['Sex_male'] + .5) * (test['Pclass'] + .5)
test['Sex_pclass'] = test['Sex_pclass'].replace(.75, "A")
test['Sex_pclass'] = test['Sex_pclass'].replace(1.25, "B")
test['Sex_pclass'] = test['Sex_pclass'].replace(1.75, "C")
test['Sex_pclass'] = test['Sex_pclass'].replace(2.25, "D")
test['Sex_pclass'] = test['Sex_pclass'].replace(3.75, "E")
test['Sex_pclass'] = test['Sex_pclass'].replace(5.25, "F")

test.head()
```

```python
## create variable of columns to drop 
to_drop = ['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked']
test = test.drop(to_drop, axis = 1)
test.head()
```

```python
## Get dummies for Pclass, Prefix, Sex_pclass
test = pd.get_dummies(test, columns= ['Pclass', 'Prefix', 'Sex_pclass'], drop_first = True)
test.head()
```

```python
passengers = test.iloc[:, 0]
passengers.columns = ['Passenger']
passengers.head()
```

```python
test = test.iloc[:, 1:]
test.head()
```

```python
## Save dataframe and import into Model notebook
test.to_csv('test_cleaned.csv', index = False)
```
